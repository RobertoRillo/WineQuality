```python
quality_counts = df['quality'].value_counts()
```

```python
plt.figure(figsize=(8,6))
plt.bar(quality_counts.index, quality_counts, color ='blue')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.set_style('darkgrid')
```

```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
```

```python
plt.figure(figsize=(14,len(numerical_columns)*6))
for idx, feature in enumerate(numerical_columns, 1):
  plt.subplot(len(numerical_columns), 2, idx)
  sns.histplot(df[feature], kde=True)
  plt.title(f'{feature} | Skewness: {round(df[feature].skew(),2)}')
```

```python
plt.tight_layout()
plt.show()
```

```python
sns.swarmplot(x="quality", y="alcohol", data=df, palette="viridis")
plt.title('Swarm Plot of Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
sns.set_palette("Pastel1")
plt.figure(figsize=(10,6))
sns.pairplot(df)
plt.suptitle('Pairplot for dataframe')
plt.show()
```

```python
sns.violinplot(x='quality', y='alcohol', data=df, palette={'3':'lightcoral','4':'lightblue','5':'lightgreen','6':'gold','7':'lightskyblue','8':'lightpink'},alpha=0.7)
plt.title('Violin Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```
